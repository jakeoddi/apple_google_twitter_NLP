## Overview

This project employs NLP and classification methods to classify text data from tweets as having positive, negative, or neutral sentiment. Several NLP preprocessing steps are taken before a Naive-Bayes classifier, Support Vector Machine, Random Forest, and an ensemble of the three are fit to the data to make predictions.

## Business Problem

As social media platforms have increasingly become the main public forum, companies have looked for new ways to derive insights about products and companies. Twitter is one of the largest of such platforms, making it great candidate with which to search for information. Analyzing sentiment regarding products and companies serves as a relatively new way to predict the products or companies that will be successful and those that won't. In addition, in can also serve as a new line to the customer to take feedback about what people liked about a product and what they didn't. Twitter sentiment can also be used to drive investment decisions, with companies that exhibit an increase in postive sentiment over time presumed to be better investments than an equal counterpart with a negative trend in sentiment.

## Data Understanding

The dataset contains 9093 tweets sourced from 2013. It was obtained from [CrowdFlower Open Source Datasets](https://data.world/crowdflower/brands-and-product-emotions) The tweets are are about either Apple or Google or any of their respective products and services. Each tweet is labeled by a human observer to be either 'Positive Sentiment', 'Negative sentiment', 'No emotion toward brand or product', or 'I can't tell.' The third feature that comes with the dataset is a column stating what product the tweet is about, however this proved unreliable on its own.

## Data Preparation, Exploratory Data Analysis (EDA), and Feature Engineering

```python
# Read in the dataset

df = pd.read_csv('./data/tweet_product_company.csv', engine = 'python')
```

## Initial Data Cleaning

Initial data cleaning involved renaming the columns, converting all tweets the the string datatype, tokenizing the tweets, then removing stop words from those tokens. Further along, I engineer more features which require their own cleaning steps.

```python
# Rename columns to something easier to work with 

df.rename(columns = {'tweet_text': 'text', 'emotion_in_tweet_is_directed_at': 'product', 
                     'is_there_an_emotion_directed_at_a_brand_or_product':'emotion'}, inplace = True)

# Converting 'text' column to str dtype

df['text'] = df['text'].astype(str)

# Tokenize using RegEx

pattern = "([a-zA-Z]+(?:'[a-z]+)?)"
df['tokens'] = df['text'].apply(lambda x: nltk.regexp_tokenize(x, pattern))

# Make sure tokens are lowercase

df['tokens'] = df['tokens'].apply(lambda x: [i.lower() for i in x])

# Create Stopwords list

stopwords_list = stopwords.words('english')
stopwords_list += list(string.punctuation)
stopwords_list += list(range(0,10))   
stopwords_list += ['sxsw', 'mention']
```

'Sxsw' and 'mention' were most common (largest) accross all three word clouds, so I added them to stop words list given they don't add much value.

```python
# Filtering out stopwords from the tokens

df['tokens'] = df['tokens'].apply(lambda x: [i for i in x if i not in stopwords_list ])
```

### WordCloud EDA

Word clouds are a great way to visualize token frequency without using traditional methods such as barplots. In my first iteration of word clouds, the terms 'sxsw' and 'mention' were the two largest for each cloud (as stated above), so I removed them and re-ran the visualization.

```python
# Create three new dataframes, each belonging exclusively to one of the four sentiments, 
# for the purpose of examining the most common words for each sentiment

df_pos = df[df['emotion'] == 'Positive emotion']
df_neg = df[df['emotion'] == 'Negative emotion']
df_neu = df[df['emotion'] == ('No emotion toward brand or product' or "I can't tell")]

# Creating list of all tokens associated with positive sentiment, then using NLTK's FreqDist to rank them by frequency

pos_tokens_list = []
for i in df_pos['tokens']:
    pos_tokens_list += i
pos_freq_dist = FreqDist(pos_tokens_list)

# Creating list of all tokens associated with negative sentiment, then using FreqDist to rank them by frequency

neg_tokens_list = []
for i in df_neg['tokens']:
    neg_tokens_list += i
neg_freq_dist = FreqDist(neg_tokens_list)

# Creating list of all tokens associated with neutral sentiment, then using FreqDist to rank them by frequency

neu_tokens_list = []
for i in df_neu['tokens']:
    neu_tokens_list += i
neu_freq_dist = FreqDist(neu_tokens_list)

# How many words we want in each wordcloud

k_most_common = 40
```

```python
# Convert the k most common words from the frequency distribution to a dictionary
# so it can be passed through the WordCloud object

pos_mc_dict = dict(pos_freq_dist.most_common(k_most_common))
pos_wordcloud = WordCloud().generate_from_frequencies(pos_mc_dict)

plt.figure(figsize = (10,10))
plt.imshow(pos_wordcloud)
plt.axis('off')
plt.show()
```

```python
neg_mc_dict = dict(neg_freq_dist.most_common(k_most_common))
neg_wordcloud = WordCloud().generate_from_frequencies(neg_mc_dict)

plt.figure(figsize = (10,10))
plt.imshow(neg_wordcloud)
plt.axis('off')
plt.show()
```

```python
neu_mc_dict = dict(neu_freq_dist.most_common(k_most_common))
neu_wordcloud = WordCloud().generate_from_frequencies(neu_mc_dict)

plt.figure(figsize = (10,10))
plt.imshow(neu_wordcloud)
plt.axis('off')
plt.show()
```

Next, I wanted to examine the sentiment based solely on the company, so I created the barplot below.

#### Text Length vs Sentiment

I wondered if there was a connection between how long a tweet was and whether it was positive, negative, or neutral. Before that, I wanted to examine the distributions of different variations of tweet-text. I created a KDE plot comparing the mean length of the raw tweets with the mean length of the tokenized tweets to observe the distributions on their own, as well as compared with one another. I then created a third KDE plot to observe the distribution of the length of the tokenized tweets.

```python
# Joining the tokens together to create a string of linked tokens for each tweet

df['tokens_linked']=[" ".join(token) for token in df['tokens'].values]

# Creating new columns to measure the length of each category I want to observe

df['raw_text_len'] = df['text'].apply(len)
df['tokens_linked_len'] = df['tokens_linked'].apply(len)
df['tokens_list_len'] = df['tokens'].apply(len)

# Calculating the mean of each new column

mean_raw_text_len = df['raw_text_len'].mean()
mean_tokens_linked_len = df['tokens_linked_len'].mean()
mean_tokens_list_len = df['tokens_list_len'].mean()
```

```python
fig, ax = plt.subplots(figsize = (9,6))
sns.kdeplot(df['raw_text_len'], ax = ax, label = 'Raw Text Length', shade = True, color = 'grey')
sns.kdeplot(df['tokens_linked_len'], ax = ax, label = 'Tokens-Linked Length', shade = True, color = 'green')
ax.set(xlabel = 'Text Length (Characters)', title = 'Text Length Density')
ax.axvline(x = mean_raw_text_len, ymax = .87, color = 'grey')
ax.axvline(x = mean_tokens_linked_len, ymax = .95, color = 'green')
ax.text(x = mean_raw_text_len, y = -.0008, s = str(mean_raw_text_len.round(1)))
ax.text(x = mean_tokens_linked_len+1, y = -.0008, s = str(mean_tokens_linked_len.round(1)))
ax.legend()
plt.show()
```

The above plot shows the distribution of raw text length to be somewhat left-skewed, with a mean of 105 characters denoted by the grey line. The distribution of the linked tokens is more uniform, skewing only very slightly to the left, and with a mean of 75 characters denoted by the green line.

```python
fig, ax = plt.subplots(figsize = (9,6))
sns.kdeplot(df['tokens_list_len'], ax = ax, label = 'Length of Tokens List', shade = True, color = 'navy')
ax.set(xlabel = 'Token List Length (Tokens)', title = 'Token List Length Density')
ax.axvline(x = mean_tokens_list_len, ymax = .95, color = 'navy')
ax.text(x = mean_tokens_list_len, y = -.005, s = str(mean_tokens_list_len.round(1)))
ax.legend()
```

The length of the original token lists seems to be uniformly distributed with a mean of 12.3 tokens denoted by the blue line.

## Modeling

With the data cleaned and explored and with the features engineered, I proceeded to model the data. First, I split up my data into train and test sets. Then I vectorized my tokens using TF-IDF vectorization, which calculates a score for each word based on its scarcity within each body of text as well as how many texts it appears in. In order for TfidfVectorizer to work on a dataframe, I needed to use my linked tokens instead of the regular list of tokens. Ready to begin modeling, I began with a Support Vector Machine, as they are suited to do well with data with many features and sparse data like text. The shape of the training data is (6819, 8078)

```python
# Assigning my predictor and target variables

data = df['tokens_linked']
target = df['emotion']

# Splitting data into train and test sets

X_train, X_test, y_train, y_test = train_test_split(data, target, random_state = 0)

# Vectorizing my train and test sets with TfidfVectorizer

vectorizer = TfidfVectorizer()
tf_idf_data_train = vectorizer.fit_transform(X_train)
tf_idf_data_test = vectorizer.transform(X_test)
```

### Support Vector Machine

```python
# Instantiating the Support Vector Machine classifier

svm = SVC()

# Then I fit the model to the training data

svm.fit(tf_idf_data_train, y_train)

# I then predict on the train and test sets

svm_train_pred = svm.predict(tf_idf_data_train)
svm_test_pred = svm.predict(tf_idf_data_test)
```

The purpose of predicting on both the train and test sets is to diagnose an imbalance between bias and variance. High performance on the training set with low performance on the test set indicates the model is overfitting to the training set and has too much variance. This can be remedied with regularization. On the other hand, if the model is underfit and has high bias, which is harder to determine, then creating more features will help.

### Naive-Bayes

```python
# Instantiating the Naive-Bayes classifier

nb = MultinomialNB()

# Fitting the model to the training data

nb.fit(tf_idf_data_train, y_train)

# Predicting on the train and test sets

nb_train_pred = nb.predict(tf_idf_data_train)
nb_test_pred = nb.predict(tf_idf_data_test)
```

### Random Forest

Because there is class imbalance within my target data, I use the 'balanced' class_weight parameter of the Random Forest classifier. Furthermore, I noticed when training that my RFC was overfitting to the training data, so I increased the number of trees to 500 from the default 100.

```python
# Instantiating the Random Forest classifier

rfc = RandomForestClassifier(n_estimators = 500, class_weight = 'balanced')

# Fitting the model to the training data

rfc.fit(tf_idf_data_train, y_train)

# Predicting on the train and test sets

rfc_train_pred = rfc.predict(tf_idf_data_train)
rfc_test_pred = rfc.predict(tf_idf_data_test)
```

## Model Evalutation

F1-score is used as the primary evaluation metric due to the class imbalance in the target variable. The SVM performed the best of the three, with an F1-score of 0.6579 on the test set. Next was the RFC with 0.6391, then the NB with 0.5922. In general, these scores are nowhere near perfect, however they do demonstrate reasonable performance accross the three models. 

```python
print('Support Vector Machine train F1 score:', f1_score(y_train, svm_train_pred, average = 'weighted'))
print('Support Vector Machine test F1 score:', f1_score(y_test, svm_test_pred, average = 'weighted'), '\n')

print('Naive-Bayes train F1 score:', f1_score(y_train, nb_train_pred, average = 'weighted'))
print('Naive-Bayes test F1 score:', f1_score(y_test, nb_test_pred, average = 'weighted'), '\n')

print('Random Forest train F1 score:', f1_score(y_train, rfc_train_pred, average = 'weighted'))
print('Random Forest test F1 score:', f1_score(y_test, rfc_test_pred, average = 'weighted'), '\n')
```

## Ensemble

Finally, I thought it would be interesting to test the performance of an ensemble of the three models. Furthermore, I wanted to cross-validate my F1-scores to improve their accuracy in measuring performance.

```python
# Fit our data with TfidfVectorizer. All of the data is fit and transformed as we are using cross validation

tf_idf_data = vectorizer.fit_transform(data)

# Instantiate the voting classifier with each of the three models. Voting is set to hard

ensemble = VotingClassifier(estimators = [('mnb', nb), ('svm', svm), ('rfc', rfc)],
                            voting = 'hard',
                            n_jobs = -1,
                            verbose = True
                           )

# Fitting the model to the training data

ensemble.fit(tf_idf_data_train, y_train)

# Predicting on the train and test data

ensemble_train_pred = ensemble.predict(tf_idf_data_train)
ensemble_test_pred = ensemble.predict(tf_idf_data_test)

# Instantiating the cross validator with the ensemble of models, data, target, and scoring metric

cv = cross_validate(estimator = ensemble, 
                    X = tf_idf_data, 
                    y = target,
                    scoring = 'f1_weighted',
                    n_jobs = -1
                   )
```

```python
print('Ensemble Cross-Validation F1 score:', cv['test_score'].mean())
```

The ensemble method performed worse on the test set (F1-score = 0.6387) than the SVM did; however, this could be due to the fact that I used cross validation with five folds on the ensemble but not the SVM.
